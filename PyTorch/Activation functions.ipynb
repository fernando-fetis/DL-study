```python
import torch
import matplotlib.pyplot as plt

fc = ['ELU', 'GELU', 'SELU', 'CELU', 'SiLU', 'ReLU']

x = torch.linspace(-10, 10, 100)
plt.figure(figsize=(20, 10))

for f in fc:
    y = eval('torch.nn.' + f)()(x)
    plt.plot(x, y, label=f, alpha=0.6)

plt.axhline(0, linestyle='--', color='k')
plt.legend(loc='center')
plt.show()
```
