```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 100)
y = np.cos(x)
```

```python
f, ax = plt.subplot_mosaic('''AAA
                              BBC
                              BBC''')

ax['A'].plot(x, y)
ax['B'].plot(x, y)
ax['C'].plot(x, y)

plt.tight_layout()
```

```python
f, ax = plt.subplots(nrows=2, ncols=3, figsize=(10,5))
ax[0, 1].plot(x, y)
# si nrows=1 o ncols=1, ax viene con squeeze=True (se elimina la dimensin adicional y los axes vienen en forma de lista plana).
```

```python
f, ax = plt.subplots(nrows=2, ncols=3, figsize=(10,5))  # retorna una tupla (desacoplable).
# ax = [[ax1, ax2, ax3], [ax4, ax5, ax6]], se puede aplanar para usar solo un ndice.
ax = ax.flatten()
ax[4].plot(x, y)
```

```python
f = plt.figure(figsize=(10, 5))

for n in range(6):
    ax = f.add_subplot(2, 3, n + 1)
    ax.plot(x, y)
```

```python
f = plt.figure()

for n in range(6):
    ax = plt.subplot(2, 3, n + 1)  # retorna un eje (no desacoplable).
    ax.plot(x, y)
    
# este mtodo es alternativo pero se prefiere plt.subplots.
```

* plt.plot(...) o plt.xlabel(...) modifica la figura actual (versin global).
* ax.plot(...) o ax.set_xlabel(...) modifica un axes en especfico.

So why do we prefer Axes?

It is refactorable - you can put away some of the code into a function that takes an Axes object, and does not rely on global state
It is easier to transition to a situation with multiple subplots
One consistent/familiar interface instead of switching between two
The only way to access the depth of all features of matplotlib

The global state version was created that way to be easy to use interactively, and to be a familiar interface for Matlab users, but in larger programs and scripts the points outlined here favour using the Axes interface.

We can for example always ask for the current axes: ax = plt.gca() ("get current axes").
